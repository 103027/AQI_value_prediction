```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import ElementClickInterceptedException
import time
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from webdriver_manager.firefox import GeckoDriverManager
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
driver.get("https://www.iqair.com/world-air-quality-ranking")
time.sleep(5)
```

```python
city = []
date = []
Time = []
aqi = []
humidity = []
wind = []
weather = []
pm25 = []
pm10 = []
no2 = []
so2 = []
o3 = []
co = []
```

```python
try:
    href = []
    html_content = driver.page_source
    soup = BeautifulSoup(html_content, 'html.parser')
    table = soup.find('div', class_='table-wrapper over-x-auto ng-star-inserted')
    table_body = table.find('tbody')
    table_rows = table_body.findAll('tr')
    for row in table_rows:
        table_data = row.findAll('td')
        a_tag = table_data[2].find('a', class_='city-label')
        href.append(a_tag['href'])
    
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
try:
    for href_ in href:
        humidity_ = ''
        wind_ = ''
        weather_ = ''
        pm25_ = ''
        pm10_ = ''
        no2_ = ''
        so2_ = ''
        o3_ = ''
        co_ = ''
        driver.get("https://www.iqair.com"+ href_)
        time.sleep(5)
        html_content = driver.page_source
        soup = BeautifulSoup(html_content, 'html.parser')
        # Extract City name , time and Date
        title_element = soup.find('div', class_='page-title')
        if title_element:
            city_name = title_element.find('h1')
            city_name = city_name.text.replace('Air quality in ', '').strip()
            date_and_time = title_element.find('h2')
            match = re.search(r'(\d{2}:\d{2}),\s*(\w+ \d+)', date_and_time.text)
            if match:
                time_ = match.group(1)  # Extracted time
                date_ = match.group(2)  # Extracted date
                current_year = datetime.now().year
                date_obj = datetime.strptime(f"{date_} {current_year}", "%b %d %Y")
                formatted_date = date_obj.strftime("%d %B %Y")
                print("Time:", time_)
                print("Date:", formatted_date)
                Time.append(time_)
                date.append(formatted_date)
            else:
                print("No time and date found.")
            print('City:',city_name)
            city.append(city_name)

        # Extract aqi
        aqiBox = soup.find('div', class_='aqi-value-wrapper__container-wrapper')
        aqiBox = aqiBox.find('p', class_='aqi-value__estimated')
        aqi_ = aqiBox.text
        print('aqi:',aqi_)
        aqi.append(aqi_)

        # Extract ,wind , weather , humidity
        weather_ = soup.find('div', class_='air-quality-forecast-container-weather')
        weather_ = weather_.text.strip()
        print(weather_)
        wind_ = soup.find('div', class_='air-quality-forecast-container-wind')
        wind_ = wind_.text.strip()
        print(wind_)
        humidity_ = soup.find('div', class_='air-quality-forecast-container-humidity')
        humidity_ = humidity_.text.strip()
        print(humidity_)

        # Extract Pollutants
        pollutants = soup.find('div', class_='pollutants-wrapper component-card full-height')
        pollutants = pollutants.find_all('div', class_='card-wrapper')
        for pollutant in pollutants:
            title = pollutant.find('div', class_='card-wrapper-info__title').text.strip()
            value = pollutant.find('span', class_='measurement-wrapper__value').text.strip()
            if title == 'PM2.5':
                pm25_ = value
                print('PM2.5',pm25_)
            elif title == 'PM10':
                pm10_ = value
                print('PM10',pm10_)
            elif title == 'O₃':
                o3_ = value
                print('O₃',o3_)
            elif title == 'NO₂':
                no2_ = value
                print('NO₂',no2_)
            elif title == 'SO₂':
                so2_ = value
                print('SO₂',so2_)
            elif title == 'CO':
                co_ = value
                print('CO',co_)
        print('----------------------------------------')
        humidity.append(humidity_)
        wind.append(wind_)
        weather.append(weather_)
        pm25.append(pm25_)
        pm10.append(pm10_)
        no2.append(no2_)
        so2.append(so2_)
        o3.append(o3_)
        co.append(co_)
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
driver.quit()
```

```python
df = pd.DataFrame({'City':city,
                   'Date':date,
                   'Time':Time,
                  'Wind':wind,
                   'Humidity':humidity,
                   'Weather':weather,
                   'PM2.5':pm25,
                   'PM10':pm10,
                   'O₃':o3,
                   'NO₂':no2,
                   'SO₂':so2,
                   'CO':co,
                   'AQI':aqi})
```
